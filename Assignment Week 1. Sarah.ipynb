```python
import seaborn as sns  
import pandas as pd  
import matplotlib.pyplot as plt 

# To make a plot of time series, I want to get the mean steps per day over all participants, need only data from steps.csv
# data pre-processing steps,get the head() of the resulting data frame

df1 = pd.read_csv('steps.csv',sep = ';') 
df1.head()
```

```python

# Removing "id" to get the mean steps  
df_steps = df1.drop("id",axis=1)
df_mean_steps = df_steps.mean(axis = 0)

print(df_mean_steps)
```

```python
# Because there are no column names on the above output, this will result in no line plot. therefore i will make a new dataframe and give it column names i will Output from above has no column name.
date = pd.to_datetime(df_mean_steps.index,format='%d-%m-%Y')
mean_steps = pd.DataFrame({'date':date,'steps':df_mean_steps})
print(mean_steps)
```

```python
# creating a plot of a time series of mean steps per day of all participants
sns.lineplot(data=mean_steps, x="date", y="steps")
```

```python
# An analysis of mean steps per day of participants : 
# Make a histogram, show the median, mean and variance value.
#somehow i cant seem to see the median 

mean_steps['steps'].name = 'Mean Steps' #Label of the variable for plotting
steps_per_day = mean_steps['steps'].dropna() #drop the NANs values 

median_mean_steps = mean_steps['steps'].median()
mean_mean_steps = mean_steps['steps'].mean()
variance_mean_steps = mean_steps['steps'].var()
print(f'The median of mean steps per day is: {median_mean_steps} ') 
print(f'The mean of mean steps per day is: {mean_mean_steps} ') 
print(f'The variance of mean steps per day is: {variance_mean_steps} ') 

sns.distplot(steps_per_day, kde=False) 
plt.title('Mean steps per day of participants')

# Adds a vertical line at x-position median_size and x-position mean_size, from y = 0 to y = 100 
plt.axvline(median_mean_steps, 0, 100, color='red', label='median') 
plt.axvline(mean_mean_steps, 0, 100, color='yellow', label='mean')   
plt.legend() #This adds a legend of the mean and median. 
plt.show()
```

```python
# Remove the values of mean steps for people with less than 2000 mean steps per day.
#print(mean_steps.sort_values(by='steps', ascending=False))

mean_steps['steps'] = mean_steps['steps'].apply(pd.to_numeric, errors='coerce')
# create a function to remove the unnecessary values,then apply to the entire column
def check_steps(x):
    if(x > 20000): 
        return float('NaN')
    else: 
        return x
mean_steps['steps'] = mean_steps['steps'].apply(check_steps)
print(mean_steps.sort_values(by='steps', ascending=False))
```

```python
# Make a histogram with new data, show the median, mean and variance value.

mean_steps['steps'].name = 'Mean Steps' #Labeling the variable for plotting
steps_per_day = mean_steps['steps'].dropna() #drop the NAN values

median_mean_steps = mean_steps['steps'].median()
mean_mean_steps = mean_steps['steps'].mean()
variance_mean_steps = mean_steps['steps'].var()
print(f'The median of mean steps per day is: {median_mean_steps} ') 
print(f'The mean of mean steps per day is: {mean_mean_steps} ') 
print(f'The variance of mean steps per day is: {variance_mean_steps} ') 

sns.distplot(steps_per_day, kde=False) 
plt.title('Mean steps per day of participants')

# Adds a vertical line at x-position median_size and x-position mean_size, from y = 0 to y = 100 
plt.axvline(median_mean_steps, 0, 100, color='green', label='median') 
plt.axvline(mean_mean_steps, 0, 100, color='red', label='mean')   
plt.legend() #This adds a legend of the mean and median. 
plt.show()
```

```python
# Make a density plot, showing a smooth distribution and provides a model of the underlying distribution of mean steps.
# Adding ticks to the actual values by using rug=True. 

sns.distplot(steps_per_day, rug=True)
plt.title('Mean steps per day')
plt.xlabel('Mean steps')
plt.ylabel('Proportion')
plt.show()
```

```python
# a violin plot representation
mean_steps = mean_steps[mean_steps['steps'] > 0]
sns.violinplot(mean_steps['steps'])
sns.swarmplot(mean_steps['steps'], color='purple') 
plt.title('Mean steps per day')
plt.ylabel('Frequency')
plt.xlabel('Mean steps')
plt.show()
```

```python
# Creating a boxplot for the mean steps
#some values fell outside the trend due to some errors in the data or maybe the data processes i used
sns.boxplot(mean_steps['steps'])
plt.title('Mean steps per day')
plt.xlabel('Mean steps')
plt.show()
```
